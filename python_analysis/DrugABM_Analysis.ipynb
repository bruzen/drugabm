```python
import pandas as pd
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
import plotly.graph_objs as go
import seaborn as sns
init_notebook_mode()

# drugOutput = pd.read_csv("model_output/output_data.txt")
# drugMapping = pd.read_csv("model_output/param_map.txt")
# drugOutput = pd.read_csv("model_output/output_data_2.txt")
# drugMapping = pd.read_csv("model_output/param_map_2.txt")

drugOutput = pd.read_csv("../output/DrugABMOutput.2016.Mar.25.20_56_59.txt")
drugMapping = pd.read_csv("../output/DrugABMOutput.2016.Mar.25.20_56_59.batch_param_map.txt")
```

```python
# To identify which runs are repeats, name each run based on parameter value mapping
runNames = []
for r in range (len(drugMapping.index)):
    runNames.append('Dealers: %i, Police: %i' %(drugMapping["initNoDealers"][r], drugMapping["initNoPolice"][r]))

# Add a new column 'runNames' to the output dataframe
drugOutput['runName'] = '' #np.zeros(len(drugOutput.index))
for r in range (len(drugOutput.index)):
    drugOutput.ix[r,'runName']=runNames[drugOutput['run'][r]-1]
```

```python
sns.tsplot(data=drugOutput, time="tick",  condition="runName", unit="run", value="noSold")
```

```python
# Show multiple confidence intervals
sns.tsplot(data=drugOutput, time="tick",  condition="runName", unit="run", value="noSold", ci=[68, 95])
```

```python
# Plot bootstrap traces
sns.tsplot(data=drugOutput, time="tick",  condition="runName", unit="run", value="noSold", err_style="boot_traces", n_boot=500)
```
